## About DS pipeline
`CRoss Industry Standard Process for Data Mining (CRISP-DM)` – стандарт, описывающий общие процессы и подходы к аналитике данных. Он является широко используемой моделью жизненного цикла проекта, которая помогает специалистам по анализу данных и бизнес-аналитикам определить необходимые этапы и процессы для успешного выполнения проекта. Этапы выглядять следующим образом:



CRISP-DM состоит из шести основных этапов, каждый из которых включает несколько подэтапов. Эти этапы включают:

- `Business Understanding`: этот этап включает определение целей и задач проекта, а также определение входных данных и ограничений.
- `Data Understanding`: на этом этапе производится предварительный анализ данных, включая их описание, проверку качества и исследование взаимосвязей.
- `Data Preparation`: на этом этапе данные подготавливаются для анализа, включая их очистку, преобразование и интеграцию.
- `Modeling`: на этом этапе создаются и проверяются модели на основе подготовленных данных.
- `Evaluation`: на этом этапе проводится оценка моделей и выбор лучшей модели.
- `Deployment`: на этом этапе разрабатывается и внедряется решение на основе выбранной модели.

# 1. Business Understanding.

Кредит стал очень важной частью жизни для многих людей и компаний. Кредиты позволяют людям и компаниям получать доступ к финансовым ресурсам, которые они могут использовать для достижения своих целей. У людей разные мотивы обращения за кредитом: кто-то может захотеть купить дом, автомобиль или даже начать бизнес, или получить личный заем. **Нехватка денег** - это распространенное предположение, которое люди делают, почему кто-то обращается за кредитом, в то время как многочисленные исследования показывают, что это не так. Даже состоятельные люди предпочитают брать кредиты, а не тратить ликвидные наличные деньги, чтобы быть уверенными, что у них достаточно резервных средств на экстренные нужды.

Банки и кредиторы получают прибыль только тогда, когда конкретный кредит погашен и не является просроченным. Когда заемщик не погашает кредит более определенного количества дней, кредитное учреждение считает, что этот кредит списан.

`Риск дефолта` – это риск того, что заемщик не будет в состоянии выплатить кредитную задолженность в полном объеме или вовсе не выплатит ее. Предсказание риска дефолта важно для банков и других кредиторов, так как позволяет им определить вероятность того, что заемщик не сможет выплатить кредит, и принять соответствующие меры для уменьшения риска убытков.

В процессе выдачи кредита банк должен оценить кредитный риск заемщика, чтобы определить, насколько вероятно, что он сможет выплатить кредит. Для этого банк анализирует кредитную историю заемщика, его доходы, трудовую историю, текущие долги и другие факторы, которые могут повлиять на его способность выплатить кредит. На основе этого анализа банк определяет кредитный рейтинг заемщика и устанавливает условия кредита, такие как процентная ставка и срок кредита.

Предсказание риска дефолта позволяет банкам принимать более обоснованные решения при выдаче кредитов. Если банк считает, что заемщик имеет высокий риск дефолта, он может отказать в выдаче кредита или установить более жесткие условия кредитования, такие как более высокую процентную ставку или более короткий срок кредита. Это позволяет банкам защитить свой капитал и уменьшить риски убытков.

Понимание бизнеса выдачи кредита важно для банков, чтобы оптимизировать процесс выдачи кредита и минимизировать риски. Для этого банки могут использовать различные методы анализа кредитного риска, такие как разработка алгоритмов оценки кредитоспособности заемщиков, а также улучшение процесса сбора и анализа информации о заемщиках.

### 2.2 Data Description

Data Description относится к процессу предоставления подробного обзора анализируемых данных. Это может включать информацию о формате, структуре, содержании, качестве данных и любых других соответствующих характеристиках, которые могут повлиять на их анализ. Это является важным этапом в любом проекте, связанном с анализом данных, так как четкое понимание данных, с которыми вы работаете, позволяет более эффективно планировать и реализовывать проект.


```python
def check_missing(data):    
    
    result = pd.concat([data.isnull().sum(),round(data.isnull().sum()/data.shape[0],4) * 100],axis=1)
    result = result.rename(index=str,columns={0:'total missing',1:'percentage'})
    
    result = result[
            result.iloc[:,1] != 0].sort_values(
        'percentage', ascending=False).round(1)
    
    print ("В выбранном датафрейме " + str(data.shape[1]) + " столбцов.\n"      
            "Всего " + str(result.shape[0]) +
              " столбцов с неполными данными.")
    
    return result[:50]

check_missing(df)
```

Датафрейм состоит из 12 столбцов(фичи) из них 2 столбца с неполными данными. Пока не будем заполнять пропуски, будем заполнять пропуски внутри пайплайна. 

Когда в данных присутствуют пропущенные значения, то их необходимо заполнить, чтобы избежать искажений в результатах анализа. Существует несколько методов заполнения пропусков:

- Удаление строк или столбцов с пропущенными значениями: Этот метод может быть использован, если количество пропущенных значений незначительно, и удаление данных не повлияет на анализ. Однако, это может привести к потере важной информации и сокращению размера выборки, что может снизить качество модели.

- Заполнение средними или медианными значениями: Этот метод может быть использован, если пропущенные значения представляют собой числовые данные. Значения могут быть заполнены средним или медианным значением соответствующего столбца. Это позволяет сохранить размер выборки и не потерять важную информацию. Однако, этот метод может исказить распределение данных и привести к ошибкам в анализе.

- Интерполяция: Этот метод может быть использован, если пропущенные значения имеют последовательную природу, например, временные ряды. Значения могут быть заполнены на основе значений, которые находятся перед и после отсутствующих значений. Этот метод может помочь сохранить свойства данных и уменьшить искажение распределения.

- Заполнение пропусков на основе ближайших соседей ( k-means): Этот метод может быть использован, если пропущенные значения представляют собой числовые или категориальные данные.Он выбирает k ближайших точек данных и использует их значения для заполнения пропусков. Для каждого пропущенного значения можно выбрать несколько ближайших соседей, используя метрику расстояния, и заполнить пропущенное значение средним или медианным значением среди этих соседей. Этот метод может быть полезен, если данные имеют множество категориальных признаков.

- Регрессионный анализ: этот метод используется, когда есть связь между пропущенными значениями и другими переменными. Модель регрессии может использоваться для заполнения пропусков, используя значения других переменных как предикторы.

Целевая переменная несбалансирована. Несбалансированная классификация - это ситуация, когда количество объектов в разных классах неравномерно распределено. По круговой диаграмме видно что объекты отрицательного класса (0) состовляют 78,2% а положительного класса (1) 21,8%. В идеале соотношение 50/50. В такой ситуации алгоритмы машинного обучения могут работать неэффективно или давать искаженные результаты, поскольку они склонны предсказывать более часто встречающийся класс. 

Способы борьбы с дисбалансом классов:
- Взвешивание классов - придадим объектам редкого класса больший вес, вместо равномерной функции потерь можно использовать взвешенную функцию потерь, чтобы учитывать неравномерное распределение классов.
- upsampling(увеличение выборки) - можно увеличить размер меньшего класса путем копирования или генерации новых примеров. Это поможет уравнять количество примеров в обоих классах и улучшить точность классификации. 
- downsampling(уменьшение выборки) - можно также уменьшить количество примеров в большем классе, чтобы сбалансировать обучающую выборку.

(существуют алгоритмы, которые специально разработаны для работы с несбалансированными данными. Некоторые из них включают в себя SMOTE, ADASYN, Random Under Sampling и т.д. Эти алгоритмы могут создавать новые примеры редкого класса, или удалять примеры частого класса, чтобы уравнять количество примеров в каждом классе.)

- Изменение порога решения(treshold): В задачах бинарной классификации можно изменить порог принятия решения, чтобы балансировать точность и полноту модели для каждого класса. Это может помочь увеличить точность для редкого класса.

- Использование метрик оценки качества: Метрики, такие как *ROC-AUC (Area Under the Receiver Operating Characteristic Curve)* и *Precision-Recall curve*, могут быть использованы для оценки качества модели для каждого класса и балансирования точности и полноты.


Для нашей модели будем использовать метрику качества классификации ROC-AUC, так как он устойчив к дисбалансу классов. О других метриках качества поговорим по-позже.

### 2.3 Data Visualisation

*Data visualization* позволяет представлять данных в графической форме для анализа и визуализации результатов. Визуализация данных позволяет наглядно и быстро представить информацию и сделать выводы. Она помогает визуализировать не только отдельные числа и статистические показатели, но и их взаимосвязь, что делает ее особенно полезной для анализа сложных данных. Визуализация может помочь в выявлении скрытых закономерностей и тенденций в данных, которые могут быть упущены при анализе только числовых значений.

**Univariate** и **bivariate** анализ данных являются основными методами визуализации данных. Оба метода позволяют анализировать и понимать данные, но они различаются по количеству переменных, которые используются в анализе.

Univariate анализ данных означает анализ только одной переменной. Это может быть переменная какого-то числового показателя, категориальная переменная или любой другой тип переменной. Univariate анализ может включать описательные статистики, такие как среднее значение, медиана, минимальное и максимальное значения, дисперсия и т. д. Эти статистические метрики могут быть представлены в виде гистограмм, ящиков с усами, диаграмм размаха, круговых диаграмм и других видов графиков(histograms, box plots, pie charts, bar charts), которые помогают визуализировать распределение данных.

Bivariate анализ данных, с другой стороны, означает анализ двух переменных одновременно. Это может быть сравнение двух числовых переменных или двух категориальных переменных, либо анализ связи между числовой и категориальной переменными. Для визуализации двух переменных могут использоваться диаграммы рассеяния, линейные графики, столбчатые диаграммы, boxplot(scatterplots, line graphs, bar graphs, box plots) и т. д. Bivariate анализ позволяет увидеть взаимосвязь между двумя переменными, что может быть полезно для принятия решений.



```python
def plot_stats(feature,label_rotation=False,horizontal_layout=True):
    
    """
    Функция которая принимает на вход столбец датафрейма, 
    и визуализирует зависимость между фичей и целевым перменным,
    в виде столбчатой диаграммы. 

    Для каждой фичи визуализируется количество займов и доля проблемных для каждой катеогрии.

    """


    temp = df[feature].value_counts()

    df1 = pd.DataFrame({feature: temp.index,'Количество займов': temp.values})

    # Расчет доли target=1 в категории
    cat_perc = df[[feature, 'loan_status']].groupby([feature],as_index=False).mean()
    cat_perc.sort_values(by='loan_status', ascending=False, inplace=True)
    
    if(horizontal_layout):
        fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,6))
    else:
        fig, (ax1, ax2) = plt.subplots(nrows=2, figsize=(12,14))
    
    sns.set_color_codes("pastel")

    s = sns.barplot(ax=ax1, x = feature, y="Количество займов", data=df1)

    if(label_rotation):
        s.set_xticklabels(s.get_xticklabels(),rotation=90)
    
    s = sns.barplot(ax=ax2, x = feature, y='loan_status', order=cat_perc[feature], data=cat_perc)
    
    if(label_rotation):
        s.set_xticklabels(s.get_xticklabels(),rotation=90)
    plt.ylabel('Доля проблемных', fontsize=10)
    plt.tick_params(axis='both', which='major', labelsize=10)
    plt.show();
```

### 2.4 Смотрим на Distributions

Просмотр распределения данных является важным этапом в анализе данных, так как он помогает понять, как данные распределены, насколько переменная дисперсивна и как она может быть использована в анализе. Распределение данных отражает, как часто появляются различные значения в наборе данных. - K-ближайших соседей: В этом алгоритме машинного обучения распределение переменных может иметь большое значение при выборе метрики расстояния между объектами.

### 3.1 Feature or a bug?

При анализе данных важно определить, является ли аномалии в данных фича или баг. Данные могут быть использованы для получения ценной информации и улучшения результатов анализа, с другой стороны снизить качество анализа и влиять на результаты алгоритмов машинного обучения.

На нашем примере кроме явных аномалии, возраст и опыт работы клиента рассмотрим как аномалии и заменим их. Остальных переменных оставим как и есть. Сначала посмотрим сколько аномалии на каждом столбце. Для этого создадим функцию outlier_detect_IQR(), и для замены аномалии impute_outlier_with_avg().

Как будем определять аномалии?

*По правилу трех сигм*: в статистике существует правило трех сигм, которое утверждает, что примерно 68% значений в нормальном распределении лежат в пределах одного стандартного отклонения от среднего значения. Используя это правило, можно определить аномалии как значения, лежащие за пределами трех стандартных отклонений от среднего значения(99 процентов значении лежит в интервале 3 сигм).

Как будем обрабатывать аномалии?

О методах обработки аномалии поговорили выше. Будем использовать медианное значение для замены аномалии.

```python
def outlier_detect_IQR(data,col,threshold=3):    
    IQR = data[col].quantile(0.75) - data[col].quantile(0.25)
    Lower_fence = data[col].quantile(0.25) - (IQR * threshold)
    Upper_fence = data[col].quantile(0.75) + (IQR * threshold)
    para = (Upper_fence, Lower_fence)
    tmp = pd.concat([data[col]>Upper_fence,data[col]<Lower_fence],axis=1)
    outlier_index = tmp.any(axis=1)
    print('Количество выбросов в данных:',outlier_index.value_counts()[1])
    print('Доля выбросов:',outlier_index.value_counts()[1]/len(outlier_index))
    return outlier_index, para


def impute_outlier_with_avg(data,col,outlier_index,strategy='median'):
    data_copy = data.copy(deep=True)
    if strategy=='mean':
        data_copy.loc[outlier_index,col] = data_copy[col].mean()
    elif strategy=='median':
        data_copy.loc[outlier_index,col] = data_copy[col].median()
    elif strategy=='mode':
        data_copy.loc[outlier_index,col] = data_copy[col].mode()[0]           
    return data_copy
```

```python
index,para = outlier_detect_IQR(data=df,col='person_emp_length',threshold=5)
print('Верхняя граница:',para[0],'\nНижняя граница:',para[1])
```

### 3.4 Feature selection

**Feature engineering** - это процесс создания новых признаков (features) на основе существующих данных для улучшения качества модели машинного обучения.

Создание новых признаков необходимо по нескольким причинам:
- Улучшение качества модели: Некоторые признаки могут не содержать достаточно информации для достижения высокого качества модели. Добавление новых признаков может помочь улучшить результаты.
- Увеличение объема данных: Иногда, если количество данных недостаточно, модель может не смочь найти в данных достаточно закономерностей. Добавление новых признаков может помочь увеличить объем данных, что может улучшить качество модели.

Примеры техник feature engineering:
- Использование группировки данных: Например, можно создать новые признаки, основанные на группировке данных по определенным категориям.
- Использование статистических методов: Например, можно создать новые признаки на основе статистических показателей, таких как среднее значение, медиана, стандартное отклонение и т.д.

Напишем свой custom transformer для feature engineering. Будем создавать новые признаки на основе существующих. Использую группировку по категориям создаем признак который показывает медиану по каждой категории.

```python
from sklearn.base import BaseEstimator, TransformerMixin

class CustomFeatureEngineering(BaseEstimator, TransformerMixin):
    """
    Args:
        include_interaction (bool): True if I want to group data by category
    Returns:
        copy of the data frame
    """
    def __init__(self, include_interaction=False):
        self.interaction = include_interaction
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, X: pd.DataFrame, y=None) -> pd.DataFrame:
        X_copy = X.copy()

        if "person_income" in set(X_copy.columns):
            X_copy['monthly_income'] = X_copy['person_income']/12
        
        if "loan_amnt" in set(X_copy.columns):
            X_copy['monthly_loan'] = X_copy['loan_amnt']/12
        
        if "monthly_income" in set(X_copy.columns):
            X_copy['monthly_pay'] = X_copy['monthly_income'] - X_copy['monthly_loan']
        
        if "person_income" in set(X_copy.columns):
            X_copy['income_without_loan'] = X_copy['person_income'] - X_copy['loan_amnt']
        
        if "cb_person_default_on_file" in set(X_copy.columns):
            X_copy['cb_person_default_on_file'] = X_copy['cb_person_default_on_file']=='Y'

        if self.interaction:
            income_by_home_ownership = X_copy[['person_income', 'person_home_ownership']].groupby('person_home_ownership').median()['person_income']
            X_copy['MEDIAN_INCOME_HOME_OWNERSHIP'] = X_copy['person_home_ownership'].map(income_by_home_ownership)
                
            income_by_intent = X_copy[['person_income', 'loan_intent']].groupby('loan_intent').median()['person_income']
            X_copy['MEDIAN_INCOME_LOAN_INTENT'] = X_copy['loan_intent'].map(income_by_intent)
                
            loan_by_cr_hist_length = X_copy[['loan_amnt', 'cb_person_cred_hist_length']].groupby('cb_person_cred_hist_length').median()['loan_amnt']
            X_copy['MEDIAN_LOAN_HIST_LENGTH'] = X_copy['cb_person_cred_hist_length'].map(loan_by_cr_hist_length)
                
            loan_by_intent = X_copy[['loan_amnt', 'loan_intent']].groupby('loan_intent').median()['loan_amnt']
            X_copy['MEDIAN_LOAN_INTENT'] = X_copy['loan_intent'].map(loan_by_intent)
                
            loan_by_loan_grade = X_copy[['loan_amnt', 'loan_grade']].groupby('loan_grade').median()['loan_amnt']
            X_copy['MEDIAN_LOAN_GRADE'] = X_copy['loan_grade'].map(loan_by_loan_grade)
    
        return X_copy
```

В этом шаге определяются методы анализа данных, которые будут использоваться для создания модели, а также параметры, которые необходимо настроить, чтобы получить наилучшие результаты.

Процес моделирования состоит из следующих этапов:
- Выбор метода моделирования: на этом этапе выбираются методы анализа данных, которые будут использоваться для построения модели.
- Выбор признаков: на этом этапе выбираются признаки, которые будут использоваться для построения модели. Это может включать выбор признаков на основе статистической значимости или важности для бизнеса.
- Выбор алгоритма: на этом этапе выбирается алгоритм, который будет использоваться для обучения модели.
- Обучение модели: на этом этапе модель обучается на основе выбранных признаков и алгоритма. Это может включать разделение данных на обучающую и тестовую выборки, настройку гиперпараметров и обучение модели на обучающих данных.
- Оценка модели: на этом этапе оценивается точность модели на основе тестовых данных и других метрик.

При моделировании будем использовать модуль *pipeline* из библиотеки sklearn. Pipeline – это модуль который позволяет автоматизировать предварительные преобразования данных перед обучением модели. Он позволяет создавать цепочки обработки данных для моделей машинного обучения и автоматизировать процесс обработки данных. Для создания Pipeline в sklearn используется класс Pipeline. Конструктор этого класса принимает список кортежей, каждый из которых состоит из имени шага и объекта, реализующего этот шаг. Шаги должны следовать в порядке обработки данных. Pipeline нашей модели будет состоять из следующих шагов:
- энкодинг переменных(преобразование категориальных переменных)
- imputing(заполнение пропусков)
- преобразование данных (стандартизация и нормализация)
- модель машинного обучения

#### Почему модель плохо предсказывает на тестовом датасете?

Наша модель может предсказать хорошо на тренировочном датасете, но низкая точность на новых данных. В МО есть понятия *overfit and underfit*. Недообучение (underfitting) происходит, когда модель не смогла захватить сложности данных в обучающем наборе и показывает плохие результаты на обучающих и тестовых данных. Для устранения недообучения можно добавить больше признаков, увеличить сложность модели, увеличить количество эпох обучения и т. д. Переобучение (overfitting) происходит, когда модель слишком точно подгоняется под обучающие данные и "запомнить" обучающий набор вместо того, чтобы извлекать общие закономерности. Это приводит к тому, что модель показывает отличные результаты на обучающих данных, но плохие на тестовых данных. Для устранения переобучения можно использовать методы регуляризации, такие как L1 или L2 регуляризация, уменьшать размерность признаков, использовать дополнительные данные для обучения и т. д. 

#### Как подобрать гиперпараметры?

Гиперпараметры модели задаются перед началом обучения модели и не определяются в процессе обучения, в отличие от обычных параметров модели, которые настраиваются в процессе обучения. Гиперпараметры контролируют настройки обучения, такие как скорость обучения, количество эпох, размер сети, метод оптимизации, параметры регуляризации и т.д. Для каждой модели МО свои гиперпараметры.

Способы подбора гиперпараметров:
- Grid Search перебор всех возможных комбинаций гиперпараметров модели, чтобы найти наилучшую комбинацию. Он может быть вычислительно дорогостоящим, но гарантирует нахождение оптимальных гиперпараметров в указанном диапазоне.
- Random Search случайным образом выбирает значения гиперпараметров в указанном диапазоне, и обучает модель на каждой комбинации. Этот метод менее вычислительно затратный, чем Grid Search, и может давать хорошие результаты, но не гарантирует нахождение оптимальных гиперпараметров.
- Bayesian Optimization использует алгоритмы оптимизации, основанные на байесовском подходе, чтобы выбирать следующие значения гиперпараметров, основываясь на результате предыдущих итераций. Этот метод может быть менее вычислительно затратным, чем Grid Search, и может давать лучшие результаты, чем Random Search.

Будем подобрать гиперпараметры через Grid Search.

#### 4.1 Логистическая регрессия

Логистическая регрессия является статистическим методом, который используется прогнозирвания вероятности принадлежности объекта к определенному классу, решает задачу бинарной классификации(только два класса). Алгоритм строится на основе зависимостей между признаками объекта и целевым значением, то есть у нас есть одна зависимая(целевая) переменная и несколько независимых (фичи) переменных. На основе зависимостей алгоритм будет строит фунцию, который минимизирует функцию потерь. Для каждой фичи расчитивается весь(вклад) на оперделение целевой переменной. Алгоритм логистической регрессии использует логистическую функцию для вычисления вероятности класса. Эта функция имеет форму сигмоиды, которая ограничивает значения от 0 до 1. Значение, близкое к 0, означает, что объект относится к первому классу, а значение, близкое к 1, означает, что объект относится ко второму классу. Для обучения модели логистической регрессии используется метод максимального правдоподобия.

#### Нужна ли для логистической регрессии преобразование данных?
Логистическая регрессия использует веса (коэффициенты), которые могут быть независимыми друг от друга и иметь разный масштаб. Это означает, что при использовании нестандартизированных данных некоторые признаки могут оказывать большее влияние на модель, чем другие. Стандартизация данных позволяет масштабировать признаки таким образом, чтобы они имели среднее значение 0 и стандартное отклонение 1. Это помогает модели логистической регрессии лучше работать, так как признаки будут иметь сравнимый масштаб и коэффициенты будут более сопоставимыми друг с другом. 

#### Какие гиперпараметры подбираем?
Есть 2 гиперпараметра логистической регрессии.
- Гиперпараметр C в модели логистической регрессии в sklearn определяет силу регуляризации модели. Регуляризация - это техника, которая используется для предотвращения переобучения модели который было описано выше(когда модель слишком точно подгоняется под обучающие данные и плохо работает на новых, не виданных ранее данных). В случае логистической регрессии гиперпараметр C определяет обратное значение силы регуляризации, где меньшие значения C соответствуют более сильной регуляризации. Когда C маленькое, алгоритм больше сосредоточен на поиске коэффициентов, которые приведут к более простой модели, даже если это означает уменьшение точности на обучающих данных. Наоборот, большие значения C означают более слабую регуляризацию и позволяют алгоритму лучше подгоняться под обучающие данные.
- Гиперпараметр penalty используются для ограничение весов признаков. Есть определенные формулы для задачи параметров регуляризации, не будем углубляться.**penalty='l1'** применяет L1-регуляризацию, которая штрафует модель за использование большого количества признаков (некоторые из коэффициентов в весовой матрице модели зануляются). **penalty='l2'** применяет L2-регуляризацию, которая штрафует модель за использование больших значений весовых коэффициентов (коэффициенты в весовой матрице модели уменьшаются).

### 4.2 kNN

kNN (k-Nearest Neighbors) алгоритм который основана на предположении, что объекты, близкие в пространстве признаков, относятся к одному и тому же классу.  Алгоритм kNN буквально запоминает вес датасет, когда на вход подается новый объект, для нового объекта классификации ищет k ближайших соседей из обучающей выборки, основываясь на метриках расстояния (евклидовая,манхеттанская,косинусная и т.д.) между объектами в пространстве признаков. Затем, чтобы определить класс нового объекта,  использует голосование большинства среди k ближайших соседей. 

Так как kNN чувствителен к масштабу и распределению данных преобразуем данные через стандартизацию который было описано выше.

**Какие гиперпараметры подбираем?** *n_neighbors* - количество соседей, используемых для классификации (по умолчанию n_neighbors=5), *weights* - весовая функция, используемая для учета расстояний до ближайших соседей (по умолчанию weights='uniform', другой вариант weights='distance' ) *metric*- метрика расстояния для измерения расстояния между объектами (по умолчанию metric='minkowski', другие варианты включают euclidean, manhattan, cosine и т.д.). Будем использовать как метрику расстояния эвклидовое расстояние.

Алгоритм kNN прост в реализации и может давать хорошие результаты для небольших и средних наборов данных, но может быть вычислительно затратным для больших наборов данных. 

```python
param_knn = {'knn__n_neighbors': [3, 5, 7, 9],
         'knn__weights': ['uniform', 'distance']}

knn = Pipeline([
    ('scaler',StandardScaler()),
    ('knn',KNeighborsClassifier(metric='euclidean'))
])

grid_search_knn = GridSearchCV(knn, param_knn, cv=10, scoring='roc_auc', n_jobs=-1, verbose=1).fit(X_train,y_train)

print("Best Hyperparameters: ", grid_search_knn.best_params_)
print('Best score: ', grid_search_knn.best_score_)
```

### 4.3 Decision Tree

*Дерево решений* основан на иерархической структуре древовидной диаграммы, который состоит из root node, nodes and leaf.  Каждый node в дереве представляет собой признак, который разделяет данные на две или более подгруппы. Каждая подгруппа связана с узлом-потомком, и процесс разделения продолжается до тех пор, пока все данные не будут отнесены к одному из предопределенных классов(на leaf) или пока не будет достигнуто некоторое условие остановки.

Алгоритм поиска наилучшего разбиения может быть разным, но обычно используются меры информативности, такие как энтропия (Entropy), индекс Джини (Gini index). Энтропия - это мера неопределенности, которая показывает, насколько хорошо данные разбиты на разные классы в узле. Индекс Джини измеряет неоднородность данных в узле, и чем он выше, тем более неоднородными являются данные в узле. Иными словами для каждого признака расчитывается значение энотропии(берем на примере энтропии) по определенной формуле, и берем тот признак который максимально снижает значение энтропии чтобы остались pure data при разбиений.

Так как алгоритм работает сценарийным подходом преобразование данных не нужен.

**Какие гиперпараметры подберем?**  *Максимальная глубина дерева (max_depth)*   задает максимальную глубину дерева. Этот параметр ограничивает количество разбиений, которые могут быть выполнены, и предотвращает переобучение. *Минимальное количество выборок в листьях (min_samples_leaf)* задает минимальное количество выборок, которые должны быть в листьях дерева. Этот параметр помогает предотвратить переобучение и уменьшить шум в данных. *Минимальное количество выборок для разделения (min_samples_split)* - задает минимальное количество выборок, необходимых для разделения узла. Этот параметр помогает предотвратить переобучение и уменьшить шум в данных. 

Дерево решений склонен к оверфиттингу, чтобы избежать переобучения можно подбирать гипрепараметры модели. Но ограничивая эти гиперпараметры есть вероятность того что наша модель не может извлекать полезные инсайты из данных. Чтобы избежать эти проблемы рекомендуют использовать pruning(обрезка).  Она заключается в том, чтобы в определённый момент обрезать ветку, не позволяя дереву переобучаться и ещё можно сначала обучить дерево и обрезаются неинформативные ветки. 

```python
param_tree = {'tree__max_depth': [2, 3, 4, 5],
              'tree__min_samples_split': [2, 4, 6, 8],
              'tree__min_samples_leaf': [1, 2, 3, 4]}

tree = Pipeline([
    ('tree',DecisionTreeClassifier(random_state=42))
])

grid_search_tree = GridSearchCV(tree, param_tree, cv=10, scoring='roc_auc', n_jobs=-1, verbose=1).fit(X_train,y_train)

print('Best score: ', grid_search_tree.best_score_)
print('Test Accuracy: ', grid_search_tree.score(X_test, y_test))
```

Случайный лес алгоритм который работает на основе дерево решений. Основная идея алгоритма в использовании несколько деревьев решений для улучшения качества прогнозов. 
Сначала извлекаем случайную выборку данных(bootstrapping) из общего набора данных (с заменой) и построим деревьев решений на основе извлеченной выборки данных. Результат прогнозируется на новых данных путем агрегирования результатов всех деревьев(среднее при задачах регрессии,голосование за большинства при классификации).Случайный лес имеет несколько преимуществ перед обычными деревьями решений, так как он уменьшает переобучение и улучшает точность прогнозов. 

Не будем усложнять модель, так как у нас мало данных, алгоритм вычислительно затратный, стандартные алгоритмы дают хорошие предсказания. В целом основные гиперпараметры похожи с деревями решений, только будем подбирать максимальную глубину деревьев. Преобразование данных не требуется.

Гиперпараметров много, не будем усложнят модель задаем только максимальную глубину дерево и количество деревьев которые будут обучатся паралельно. При алгоритме случайного леса деревя обучаются независимо друг от друга.

### 4.5 Градиентный бустинг

Градиентный бустинг (Gradient Boosting) второй алгоритм основанный на ансамбле решающих деревьев. Он работает следующим образом:
- Создается начальная модель, которая представляет из себя простую модель(напримере задачи регрессии модель которая всегда предсказывает среднее значение целевой переменной)
- Считается ошибка между предсказаниями начальной модели и настоящими значениями целевой переменной(residuals).
- Далее, создается следующая модель, которая приближает ошибку, полученную на первом шаге. Такая модель называется базовой моделью. *Note*: работаем не с целевым пременным а ошибкой.
- Используя базовую модель, считается новая ошибка между предсказаниями на первом шаге и настоящими значениями целевой переменной.
- Создается следующая базовая модель, которая приближает новую ошибку, и процесс продолжается до тех пор, пока не будет достигнут заданный порог точности.
- На этапе предсказания новых значений для тестовых данных, каждая базовая модель применяется по очереди, и предсказания моделей суммируются. Результат этой суммы представляет собой финальное предсказание модели.

С иными словами работает как градиентный спуск(численный метод который используется для подбора весов для линейных моделей), но минимизирует ошибки предыдущего дерева. Последовательно шагами приближает ошибку модели к минимуму. 

Первый алгоритм который будем использовать для бустинга XGBoost (eXtreme Gradient Boosting). Не будем усложнять модель, будем подобрать только глубину дерево и количество деревьев как в случае алгоритма случайного леса.Через learning rate задаем шаг обучения,это значение задает насколько быстро алгоритм будет корректировать ошибки модели. Более низкий learning rate может улучшить качество модели, но также может увеличить время обучения. 

При алгоритме градиентного бустинга деревья обучаются последовательно.

```python
param_xg = {
    'model__max_depth': [3, 4],
    'model__learning_rate': [0.1, 0.01, 0.001],
    'model__n_estimators': [50, 100]
}

model = xgb.XGBClassifier()

xg = Pipeline([
    ('model',model)
])

grid_xg = GridSearchCV(xg, param_xg,cv=10, scoring='roc_auc', n_jobs=-1, verbose=1).fit(X_train,y_train)
print('Best score: ', grid_xg.best_score_)
print('Test Accuracy: ', grid_xg.score(X_test, y_test))
```

LightGBM работает на основе подхода гистограммным градиентным бустингом, который использует гистограммы для ускорения вычислений и сокращения использования памяти. ### 4.6 Voting and Stacking

*VotingClassifier* работает путем объединения прогнозов нескольких базовых алгоритмов для получения окончательного прогноза. На нашем примере работает метод голосование за большинству. В классификаторе голосования есть два типа голосования: hard голосование и soft голосование. При hard голосовании окончательный прогноз основывается на большинстве голосов базовых классификаторов(какой класс наберет больше голосов). При soft голосовании окончательный прогноз основывается на средневзвешенном значении прогнозируемых вероятностей базовых классификаторов(predict probablity а не класс). 

Будем использовать ранее созданные пайплайны для базовых алгоритмов и голосовать будем на примере soft.

```python
voting = VotingClassifier(estimators=[
    ('lr',lr),
    ('knn',knn),
    ('tree',tree)
],voting='soft')

voting.fit(X_train,y_train)
```

```python
x_pred = voting.predict(X_test)
roc_auc_score(x_pred,y_test)
```

*Стекинг* объединяет несколько различных моделей, обученных на одних и тех же данных, но с использованием разных алгоритмов, для получения лучшей точности предсказаний. Сначала делим обучающий набор на две или более частей. На первой части обучаются несколько базовых моделей с использованием различных алгоритмов машинного обучения, таких как деревья решений, логистическая регрессия, случайный лес и т.д. Затем, на второй части данных, каждая из базовых моделей делает предсказания, которые используются в качестве входных данных для обучения "мета-модели". Мета-модель обычно использует простую линейную(логистическую) регрессию или другой алгоритм машинного обучения для комбинации предсказаний базовых моделей и создания окончательных предсказаний.

В результате стекинга получается ансамбль моделей, который может иметь лучшую точность предсказаний, чем каждая из базовых моделей в отдельности. В отличий от Voting-a который использует большинство голосов для генерации окончательного прогноза, при стекинге используется мета-алгоритм который презсказывает окончательный прогноз. Как final_estimator используем логистическую регрессию.

```python
stacking = StackingClassifier(estimators=[
    ('lr',lr),
    ('knn',knn),
    ('tree',tree)
],final_estimator=LogisticRegression())

stacking.fit(X_train,y_train)
```

```python
x_pred = stacking.predict(X_test)
roc_auc_score(x_pred,y_test)
```

#### Допустим, наша модель сделал предсказания, насколько она точна?

Для проверки точности модели используются метрики качества. В зависимости задачи машинного обучения, у каждой задачи свой метрики. На нашем примере существует метрики,такие как точность, полнота,f1-мера, ROC-AUC. Мы считали точность модели через метрику ROC-AUC. Разберем какие метрики когда лучше использовать. 

При классификации есть понятие матрицы ошибок(confusion matrix). Матрица ошибок позволяет проанализировать, насколько точно классификатор предсказывает классы объектов.Матрица ошибок имеет вид таблицы, в которой по одной оси указаны истинные метки классов, а по другой - предсказанные классы. Каждая ячейка таблицы содержит число объектов, которые были классифицированы в соответствии с соответствующей комбинацией истинных и предсказанных меток классов.



Матрица ошибок состоит из четырех элементов:
- True Positive (TP) - количество объектов, которые были правильно классифицированы как положительные;
- False Positive (FP) - количество объектов, которые были неправильно классифицированы как положительные;
- False Negative (FN) - количество объектов, которые были неправильно классифицированы как отрицательные;
- True Negative (TN) - количество объектов, которые были правильно классифицированы как отрицательные.

На основе этих элементов можно вычислить вышесказанные метрики.
- Точность (precision) представляет из себя долю истинных срабатываний от общего количества срабатываний. Она показывает, насколько можно доверять алгоритму классификации в случае срабатывания

$$precision(a, X) = \frac{TP}{TP+FP}.$$

- Полнота (recall) считается как доля объектов, истинно относящихся к классу "+1", которые алгоритм отнес к этому классу

$$recall(a, X) = \frac{TP}{TP+FN},$$

здесь $TP+FN$ как раз будут вместе составлять весь список объектов класса "+1".

Точность и полнота зависят от выбранного порога классификации, поэтому при выборе порога нужно учитывать, какую метрику производительности вы хотите оптимизировать. Например,если минимизируем число ложных положительных результатов надо выбрать порог, который максимизирует точность. Если  минимизируем число ложных отрицательных результатов то надо выбрать порог, который максимизирует полноту.


Есть способы оценки модели в целом, не привязываясь к конкретному порогу. Первый из них основан на использовании _ROC-кривой_. Такая кривая строится в следующих координатах:

по оси $x$ откладывается доля ложных срабатываний (False Positive Rate) - отношение числа ложных срабатываний к общему размеру отрицательного класса:

$$FPR = \frac{FP}{FP+TN}$$

по оси $y$ откладывается доля верных срабатываний (True Positive Rate) - отношение числа верных срабатываний к размеру положительного класса:

$$TPR = \frac{TP}{TP+FN},$$

то есть TPR по сути представляет из себя полноту, о которой мы говорили ранее.

Точка на графике будет соответствовать конкретному классификатору с некоторым значением порога.
ROC кривая всегда идет из точки (0,0) в точку (1,1). При этом в случае наличия идеального классификатора с определенным порогом доля его верных ответов будет равна 1, а доля ложных срабатываний - 0, то есть график будет проходить через точку (0,1). Таким образом, чем ближе к этой точке проходит ROC-кривая, тем лучше наши оценки и лучше используемое семейство алгоритмов. Таким образом мерой качества оценок принадлежности к классу 1 может служить площадь под ROC-кривой. Такая метрика называется AUC-ROC (Area Under Curve - площадь под кривой ROC). В случае идеального алгоритма $AUC-ROC = 1$, а в случае худшего приближается к $\frac{1}{2}$(рандомность когда щансы 50/50).

Мы использовали метрику AUC-ROC для оценки качества модели так как он устойчив к treshold(порог, учитывает точность модели при разных значениях порогов) и метрика не зависит от соотношения между положительными и отрицательными примерами(у нас отношение классов 70/20).
