# Step1. Business Understanding


Главная задача любого бизнеса — зарабатывать деньги. Метрики продукта – это ключевые показатели, которые используются для оценки успеха продукта на рынке. Они помогают бизнесам измерить эффективность продукта и улучшить его пользовательский опыт. Главная задача бизнеса — зарабатывать деньги. Вот некоторые из наиболее распространенных метрик продукта:

1.	`Активность пользователей`: Это метрика, которая отслеживает количество активных пользователей продукта в течение определенного периода времени. Она может включать в себя такие показатели, как DAU (ежедневно активные пользователи), MAU (ежемесячно активные пользователи) и WAU (еженедельно активные пользователи).

2.	`Удержание пользователей`: Эта метрика отслеживает процент пользователей, которые продолжают использовать продукт после первого запуска. Она помогает понять, насколько продукт удобен и полезен для пользователей.

3.	`Конверсия`: Эта метрика отслеживает процент пользователей, которые выполняют целевое действие, такое как регистрация, подписка или покупка. Конверсия помогает оценить эффективность маркетинговых кампаний и пользовательского интерфейса продукта.

4.	`Частота использования`: Эта метрика отслеживает количество раз, когда пользователи используют продукт в течение определенного периода времени. Частота использования может быть ключевым показателем для продуктов, которые ориентированы на повторное использование.

5. `Экономика одной продажи(unit-экономика) и экономика одного покупателья(LTV, CAC)`.  **LTV (Lifetime Value)** - это показатель, который описывает, сколько денег в среднем зарабатывает компания от одного клиента в течение всего периода его сотрудничества с ней. Этот показатель позволяет оценить, насколько прибыльны клиенты в долгосрочной перспективе и сколько компания может потратить на их привлечение и удержание. **CAC (Customer Acquisition Cost)** - это показатель, который описывает, сколько денег компания тратит на привлечение одного нового клиента. Этот показатель позволяет оценить, насколько эффективно компания тратит свои ресурсы на привлечение новых клиентов и увеличение своей клиентской базы.


# Step 2. Data understanding

Мы будем анализировать данные о клиентах магазина товаров Google (также известного как GStore). Маркетинговая аналитика позволяет бизнесу оценить эффективность своих маркетинговых кампаний и каналов продвижения. Маркетинговая аналитика позволяет бизнесу оптимизировать возврат инвестиций в маркетинг. Анализ затрат на маркетинг и отслеживание метрик позволяет определить наиболее эффективные каналы и кампании, чтобы максимизировать результаты и минимизировать затраты.

`Логи` в маркетинге - это записи, которые содержат информацию о действиях пользователей взаимодействующих с рекламными и маркетинговыми материалами компании. Логи обычно содержат данные о действиях пользователей на сайте, в приложении или в других каналах маркетинга, таких как электронная почта или социальные сети.

В логах маркетинга могут содержаться различные данные, такие как:
- Данные о посещении сайта или страницы
- Данные о продуктах, которые были просмотрены или куплены
- Данные о поисковых запросах или ключевых словах
- Данные о рекламных кликах и просмотрах
- Данные о поведении на сайте или в приложении
- Данные о географическом местоположении пользователя
- Данные о девайсах, используемых пользователем

Эти данные могут быть использованы для анализа маркетинговых кампаний и определения эффективности различных маркетинговых стратегий. Например, логи могут помочь компании понять, какие каналы маркетинга наиболее эффективны для достижения конкретных целей, какие продукты и страницы на сайте наиболее популярны, и какие ключевые слова наиболее эффективны для продвижения продуктов и услуг компании.

```python
data['session_year'] = data['visitStartTime'].dt.year
data['session_month'] = data['visitStartTime'].dt.month

session_per_user = (
    data.groupby(['session_year','session_month'])
    .agg({
        'fullVisitorId': 'nunique',
        'sessionId': 'count'
    })
    .sort_values(['session_year','session_month'])
)

session_per_user.columns = [
    'n_users',
    'n_sessions'
]

session_per_user['session_per_user'] = session_per_user['n_sessions'] / session_per_user['n_users']
session_per_user
```

Оценим лояльность и вовлеченность аудитории (Sticky Factor). Sticky factor отражает регулярность использования сервиса, и для недельной аудитории рассчитывается как DAU/WAU. Для месячной аудитории — DAU/MAU.

```python
#для месячной аудитории
mau_total = (
    data.groupby(['session_year', 'session_month'])
    .agg({'fullVisitorId': 'nunique'})
    .mean()
)

#для дневной аудитории
dau_total = (
    data.groupby('session_day').agg({'fullVisitorId': 'nunique'}).mean()
)

#для недельной аудитории
wau_total = (
    data.groupby(['session_year', 'session_week'])
    .agg({'fullVisitorId': 'nunique'})
    .mean()
)


sticky_wau = dau_total/wau_total*100 
print('Sticky Factor для недельной аудитории %s' % sticky_wau[0].round(2))
sticky_mau = dau_total/mau_total*100
print('Sticky Factor для месячной аудитории %s' %sticky_mau[0].round(2))
```

# 3. Как часто возвращаються пользователи?

Анализ поведения пользователей во времени дает ценную информацию об эффективности рекламных кампаний. В частности, можно определить влияние разных маркетинговых инструментов на ключевые показатели бизнеса: LTV, конверсия, ROМI, Retention Rate, САС и т.д.

В идеале можно было расчитать метрики, но у нас нет доступных данных. С помощью данных о расходах на маркетинг можно было расчитать ROМI(Return on Marketing Investment)- коэффициент возврата маркетинговых инвестиций , CAC (customer acquisition cost) — затраты на привлечения клиента и т.д.

Будем строит когорты и анализируем метрики.Когортный анализ - это метод исследования поведения группы людей (когорты), которые имеют общую характеристику или опыт за определенный период времени. Он позволяет более точно понимать, как пользователи взаимодействуют с продуктом или услугой в разные периоды времени и как изменяется их поведение с течением времени. Сформируем когорты по первому посещению сайта.

```python
# Для каждого пользователья найдем дату первого посещения сайта, и мерджим с исходным датафреймом
first_visit = data.groupby(['fullVisitorId'])['visitStartTime'].min().reset_index()
first_visit.columns = ['fullVisitorId', 'first_visit']

data = data.merge(first_visit, on='fullVisitorId')
data.head()
```

Отформатируем по datetime:
для каждого найдем месяц первого посещения и месяц посещения

```python
data['first_visit_month'] = data['first_visit'].astype('datetime64[M]')
data['visit_month'] = data['visitStartTime'].astype('datetime64[M]')
```

```python
data.pivot_table(
    index='first_visit_month',
    columns='visit_month',
    values='fullVisitorId',
    aggfunc='nunique',
).fillna('')
```

Сформировали сводную таблицу который показывает количество вернувшихся пользователей,после первого посещения сайта. Пользователи которые посещали сайт впервые 2016-08-01 было 61126.0, за следущий месяц из них вернулись 2194.0,а за второй месяц вернулись 1032.0. Для магазина который продает футболки довольно неплохо.

Теперь будем смотреть на среднюю выручку по когортам.

```python
revenue_by_cohorts = data.groupby(
    ['first_visit_month', 'visit_month']
).agg({'totals_transactionRevenue': 'sum', 'fullVisitorId': 'nunique'})


revenue_by_cohorts.columns = ['revenue','users_count']

revenue_by_cohorts.pivot_table(
    index='first_visit_month',
    columns='visit_month',
    values='revenue',
    aggfunc='mean',
).fillna('')
```

По сводной таблице можно сделать вывод о том, что пользователи в основном заходять и покупают при первой посещений сайта, из них повторную покупку делают относительно небольшое количество пользователей.

А что с средней выручкой на каждого пользователья(ARPU)?

```python
revenue_by_cohorts['revenue_per_user'] = (
    revenue_by_cohorts['revenue'] 
    / revenue_by_cohorts['users_count']
)

revenue_by_cohorts.pivot_table(
    index='first_visit_month',
    columns='visit_month',
    values='revenue_per_user',
    aggfunc='mean',
).fillna('')
```

Видно что, за первые месяцы когорт число пользователей много но выручка на каждого пользователья относительно низкая. Хотя, показатель растет для пользователей которые вернулись после первого посещения сайта. 

Следующий шаг это расчет lifetime для каждой когорты. Нулевой lifetime показывает число первых посещений для каждой когорты. 

```python
revenue_by_cohorts.reset_index(inplace=True)
revenue_by_cohorts['cohort_lifetime'] = ((
    revenue_by_cohorts['visit_month']
    - revenue_by_cohorts['first_visit_month']
) / np.timedelta64(1, 'M')).round().astype('int')
```

```python
cohorts = revenue_by_cohorts.pivot_table(
    index='first_visit_month',
    columns='cohort_lifetime',
    values='revenue_per_user',
    aggfunc='mean',
)

cohorts
```

Для наглядности, можем строит тепловую карту.

```python
plt.figure(figsize=(13,9))
plt.title('Средний чек покупателей')
sns.heatmap(cohorts, annot=True,fmt='.2f',linewidths=1, linecolor='black',cmap="YlGnBu")
```

По тепловой карте видно, аномально высокие значения среднего чека. Пользователи делали покупку на крупную сумму, можно предположить что от магазина было сделано акция для постоянных пользователей.

## 3.1 Retention Rate

Retention rate (коэффициент удержания) - это метрика, используемая для измерения того, сколько клиентов остается у компании после определенного периода времени. Удержание клиентов обходится компаниям дешевле, чем привлечение новых клиентов. Удержание клиентов снижает расходы на маркетинг и рекламу, так как не требуется постоянно привлекать новых клиентов а так же удержание клиентов ведет к повторным покупкам, увеличению продаж и росту доходов.

```python
# Посчитаем количество пользователей для каждой когорты за 0 лайфтайм
users_count_0_lifetime = revenue_by_cohorts[revenue_by_cohorts['cohort_lifetime'] == 0][
    ['first_visit_month', 'users_count']
]
print(users_count_0_lifetime)

revenue_by_cohorts = revenue_by_cohorts.merge(users_count_0_lifetime,on='first_visit_month')
```

```python
revenue_by_cohorts['retention'] = revenue_by_cohorts['users_count_x']/revenue_by_cohorts['users_count_y']

retention_pivot = revenue_by_cohorts.pivot_table(
    index='first_visit_month',
    columns='cohort_lifetime',
    values='retention',
    aggfunc='sum',
)
retention_pivot
```

```python
sns.set(style='white')
plt.figure(figsize=(13, 9))
plt.title('Cohorts: User Retention')
sns.heatmap(retention_pivot, annot=True, fmt='.1%', linewidths=1, linecolor='gray');
```

С удержанием клиентов у нас проблемы. Видно за первый лайфтайм 2-3  процентов пользователей возвращаются, к концу почти 0-0.1 процентов. Маркетологам предстоит поработать с удержанием клиентов.

## 3.2 Churn Rate

Churn rate - это показатель, который отображает, какую долю клиентов компании перестали пользоваться ее продуктами или услугами в определенный период времени. Показатель эффективности бизнеса: Если churn rate высок, то это может свидетельствовать о том, что продукт или услуга не удовлетворяют потребности клиентов или компания не обеспечивает достаточно высокого уровня обслуживания. Высокий churn rate может означать, что компания теряет деньги и не растет, в то время как низкий churn rate свидетельствует о том, что бизнес эффективен и имеет хорошие перспективы.

```python
churn = data[['fullVisitorId','first_visit_month','visit_month']]

ch = churn.groupby(['first_visit_month','visit_month']).aggregate({'fullVisitorId':'nunique'}).reset_index()

ch['cohort_lifetime'] = ((
    ch['visit_month']
    - ch['first_visit_month']
) / np.timedelta64(1, 'M')).round().astype('int')
```

```python
# Воспользуемся встроенным методом pct_change который показывает процентное изменение значения 
ch['churn_rate'] = ch.groupby(['first_visit_month'])['fullVisitorId'].pct_change()

churn_pivot = ch.pivot_table(
    index='first_visit_month',
    columns='cohort_lifetime',
    values='churn_rate',
    aggfunc='sum',
)
churn_pivot
```

```python
sns.set(style='white')
plt.figure(figsize=(13, 9))
plt.title('Cohorts: Churn Rate')
sns.heatmap(churn_pivot, annot=True, fmt='.1%', linewidths=1, linecolor='gray')
```

С оттоком клиентов тоже стоит поработать. За первый лайфтайм 96-97 процентов пользователей не возвращаются. Из оставщихся пользователей количество убывается за каждый последующий лайфтайм.

# 4. RFM - анализ

RFM-анализ (Recency, Frequency, Monetary) является  инструментом маркетингового анализа, который позволяет классифицировать клиентов на основе их покупательского поведения и эффективно использовать эти знания для улучшения маркетинговых стратегий и увеличения прибыли компании.

С помощью RFM-анализа в том, что мы разделяем всех клиентов на группы, в зависимости от того, как давно они сделали последнюю покупку, как часто покупали и насколько большой была сумма их заказов.

 - Recency — давность (как давно ваши клиенты что-то у вас покупали);
 - Frequency — частота (как часто они у вас покупают);
 - Monetary — деньги (общая сумма покупок).
 
RFM-анализ позволяет выделить наиболее ценных клиентов, которые совершают покупки чаще, тратят больше денег и делают это недавно. RFM-анализ может помочь бизнесу определить свою целевую аудиторию, настроить маркетинговые кампании, разработать программу лояльности, а также выявить слабые места в продажах и принять меры по их устранению.

```python
# по сколько покупок есть на покупателя? Для каждого пользователя 0.01 покупок, не имеет смысла кластеризовать клиентов.
data['totals_transactionRevenue'].count() / data['fullVisitorId'].nunique()
```

```python
#для РФМ анализа оставим только тех кто-покупал
rfm_data = data[~data['totals_transactionRevenue'].isnull()]
rfm_data.reset_index(drop=True,inplace=True)
rfm_data['totals_transactionRevenue'].count() / rfm_data['fullVisitorId'].nunique()
```

Для тех кто покупал тоже показатели не очень хороши, но будем довольствоваться с этим.

```python
max_data = rfm_data['visitStartTime'].max()
max_data
```

```python
rfm= rfm_data.groupby('fullVisitorId').agg({
    #Recency
    'visitStartTime': lambda x: (max_data - x.max()).days,
    #Frequency
    'sessionId':'nunique',
    #Monetary
    'totals_transactionRevenue':'sum'
})

rfm.columns = ['Recency','Frequency','Monetary']

rfm.sort_values(by='Frequency',ascending=False)
rfm = rfm.drop(1957458976293878100)
```

# 5. Кластеризация пользователей

При кластеризации разделяем объектов на группы (кластеры), таким образом, чтобы объекты внутри одной группы были более похожи друг на друга, чем на объекты из других групп. Другими словами группируем клиентов в подмножества, которые обладают схожими характеристиками. Эти инструменты позволяют компаниям лучше понимать своих клиентов, идентифицировать потребности, желания и предпочтения, а также определять потенциальные проблемы  в поведении клиентов. На основе этой информации компании могут более эффективно настраивать свои маркетинговые и продуктовые стратегии, а также улучшать качество обслуживания своих клиентов.

Для кластеризации пользователей будем использовать алгоритм k-Means. Алгоритм k-Means работает следующим образом:
- Выбираем количество кластеров k, которое требуется создать.
- Случайным образом выберем k точек данных из множества всех точек входных данных. Эти точки станут центрами k кластеров.
- Для каждой точки данных определим ближайший к ней центр кластера и назначим эту точку к соответствующему кластеру.
- Для каждого кластера вычисляем новый центр, который будет являться средним значением всех точек в кластере.
- Повторяем шаги 3 и 4 до тех пор, пока центры кластеров не стабилизируются.


```python
from sklearn.cluster import KMeans, DBSCAN
from sklearn.preprocessing import StandardScaler
import plotly.express as px 
from sklearn.metrics import silhouette_score
```

```python
scaler = StandardScaler() # создаём объект класса scaler (нормализатор)
scaler.fit(rfm)
x_sc = scaler.transform(rfm) # обучаем нормализатор и преобразуем набор данных 
rfm_normal = pd.DataFrame(x_sc, index=rfm.index, columns=rfm.columns)
```

#### Как выбрать оптимальное количество кластеров?

Используем метод "локтя"(elbow). Этот метод строить график зависимости суммарной внутрикластерной суммы квадратов от количества кластеров и найти точку на графике, где резкий спад прекращается и график становится плавным. Эта точка считается оптимальным количеством кластеров. На нашем случае по графику видно что после 4 кластеров резкий спад прекращается. Выберем это число как количество кластеров.

```python
WCSS = []
for i in range(1,11):
    model = KMeans(n_clusters = i,init = 'k-means++')
    model.fit(rfm_normal)
    WCSS.append(model.inertia_)
    
fig = plt.figure(figsize = (7,7))
plt.plot(range(1,11),WCSS, linewidth=4, markersize=12,marker='o',color = 'green')
plt.xticks(np.arange(11))
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
# задаём модель k_means с числом кластеров 4

km = KMeans(n_clusters = 4,random_state=42)

# прогнозируем кластеры для наблюдений (алгоритм присваивает им номера от 0 до 3)
labels = km.fit_predict(rfm_normal)

# сохраняем метки кластера в поле нашего датасета
rfm['cluster_km'] = km.labels_
```

Визуализируем на 3D пространстве кластера. 

```python
import plotly.graph_objects as go

PLOT = go.Figure()

for C in list(rfm.cluster_km.unique()):
    
    PLOT.add_trace(go.Scatter3d(x = rfm[rfm.cluster_km == C]['Recency'],
                                y = rfm[rfm.cluster_km == C]['Frequency'],
                                z = rfm[rfm.cluster_km == C]['Monetary'],
                                mode = 'markers', marker_size = 8, marker_line_width = 1,
                                name = 'Cluster ' + str(C)))

PLOT.update_layout(width = 800, height = 800, autosize = True, showlegend = True,
                   scene = dict(xaxis=dict(title = 'Annual Income', titlefont_color = 'black'),
                                yaxis=dict(title = 'Age', titlefont_color = 'black'),
                                zaxis=dict(title = 'Spending Score', titlefont_color = 'black')),
                   font = dict(family = "Gilroy", color  = 'black', size = 12))
```
